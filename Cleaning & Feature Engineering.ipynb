```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("./dataset/train-data.csv")
data.head()
```

```python
# remove unnamed column
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
data.describe()
```

```python
print(data['Location'].unique())
print(data['Fuel_Type'].unique())
print(data['Transmission'].unique())
print(data['Owner_Type'].unique())
```

```python
data.isnull().sum()
```

```python
print('Shape of data before dropping any Rows: ', data.shape)
data = data[data['Mileage'].notna()]
print("Shape of train data after dropping row with null values in mileage: ", data.shape)
data = data[data['Engine'].notna()]
print("Shape of data after dropping Rows with null values in Engine: ", data.shape)
data = data[data['Power'].notna()]
print("Shape of train data after dropping rows with null values in power: ", data.shape)
data = data[data['Seats'].notna()]
print("Shape of train data After dropping Rows with null values in Seats: ", data.shape)
```

```python
data = data.reset_index(drop = True)
```

# Feature Engineering
There are many different data which could be extarcted from present. And, that's where Feature Engineering comes

```python
for i in range(data.shape[0]):
    data.at[i, 'Company'] = data['Name'][i].split()[0]
    data.at[i, 'Mileage(km/kg)'] = data['Mileage'][i].split()[0]
    data.at[i, 'Engine(CC)'] = data['Engine'][i].split()[0]
    data.at[i, 'Power(bhp)'] = data['Power'][i].split()[0]
    
data['Mileage(km/kg)'] = data['Mileage(km/kg)'].astype(float)
data['Engine(CC)'] = data['Engine(CC)'].astype(float)
```

```python
data.head()
```

At this point when I tried to change Power(bhp) to float an error occured (Can't convert str to float : null). So, I started investigating. And found some cell where having values: 'null bhp

```python
x = 'n'
count = 0
position = []
for i in range(data.shape[0]):
    if data['Power(bhp)'][i] == 'null':
        x = 'Y'
        count += 1
        position.append(i)

print(x)
print(count)
print(position)
```

```python
data = data.drop(data.index[position])
data = data.reset_index(drop = True)
data.shape
```

```python
data['Power(bhp)'] = data['Power(bhp)'].astype(float)
data.head()
```

```python
for i in range(data.shape[0]):
    if pd.isnull(data.loc[i, 'New_Price']) == False:
        data.at[i, 'New_car_Price'] = data['New_Price'][i].split()[0]

data['New_car_Price'] = data['New_car_Price'].astype(float)
```

```python
# Delete all useless features

data.drop(['Name'],axis = 1, inplace = True)
data.drop(['Mileage'], axis = 1, inplace = True)
data.drop(['Engine'], axis = 1, inplace = True)
data.drop(['Power'], axis = 1, inplace = True)
data.drop(['New_Price'], axis = 1 , inplace = True)
```

```python
data.to_csv('cleaned_data.csv', index = False)
```
